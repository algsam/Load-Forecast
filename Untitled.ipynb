```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import sqlite3
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from math import sqrt
import statsmodels.api as sm
from statsmodels.tsa import stattools
import os
```

```python
data = pd.read_csv('C:/Users/282512/LF/testss.csv', sep=',')
#delete the unwanted data
del data['25']
del data['Unnamed: 28']
Load_filter = data['VTYPE'].str.contains('LOAD')
Temp_filter = data['VID'].str.contains('TEMP')
Hum_filter =  data['VID'].str.contains('HUM')
Load = data[Load_filter]
Temp = data[Temp_filter]
Hum = data[Hum_filter]
del Load['VID']
del Load['VTYPE']
del Load['DATE']
del Temp['VID']
del Temp['VTYPE']
del Temp['DATE']
del Hum['VID']
del Hum['VTYPE']
del Hum['DATE']
load_matrix = Load.as_matrix()
Temp_matrix = Temp.as_matrix()
Hum_matrix = Hum.as_matrix()
a = load_matrix[0]
b = Temp_matrix[0]
c = Hum_matrix[0]
for i in range(1,751):
    a = np.append(a,load_matrix[i])
    b = np.append(b,Temp_matrix[i])
    c = np.append(c,Hum_matrix[i])
    
a = np.append(a,0)
b = np.append(b,0)
c = np.append(c,0)
rng = pd.date_range('06/20/2016', freq = 'H', periods = 18025, name = 'Date')
data1 = pd.Series(a, index = rng)
Tab = data1.to_frame('Load')
Tab['Temp'] = pd.Series(b, index = Tab.index)
Tab['Hum'] = pd.Series(c, index = Tab.index)
Tab1 = Tab.dropna()
filter = Tab1['Load']>100
LF = Tab1[filter]
LF = LF.reset_index() #if the index is 'Date', you can not use dt.dayofweek, therefore the index must be reset,
#ie Date is not the index anymore
LF['weekday'] = LF['Date'].dt.dayofweek
LF.set_index('Date',inplace=True)
filter_1 = LF['weekday'] > 5 
filter_2 = LF['weekday'] < 4 
filter_3 = LF['weekday'] == 5 
filter_4 = LF['weekday'] == 4 
LF_weekday = LF[filter_1 | filter_2]
LF_weekend = LF[filter_3 | filter_4]
```
